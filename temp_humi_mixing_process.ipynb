1. MAE (Mean Absolute Error) 
- 실제 값과 예측 값의 차이(error)를 절대값으로 변환해 평균화  
- MAE는 에러에 절대값을 취하기 때문에 에러의 크기 그대로 반영된다. 그러므로 예측 결과물의 에러가 10이 나온 것이 5로 나온 것보다 2배가 나쁜 도메인에서 쓰기 적합한 산식이다.
- 에러에 따른 손실이 선형적으로 올라갈 때 적합하다.
- 이상치가 많을 때  

2. MSE(Mean Squared Error)  
- 실제 값과 예측 값의 차이를 제곱해 평균화
- 예측값과 실제값 차이의 면적의 합
- 특이값이 존재하면 수치가 많이 늘어난다. 

3. RMSE(Root Mean Squared Error) 
- MSE 값은 오류의 제곱을 구하므로 실제 오류 평균보다 더 커지는 특성이 있어 MSE에 루트를 씌운 RMSE 값을 쓴다.
- 에러에 제곱을 하기 때문에 에러가 크면 클수록 그에 따른 가중치가 높이 반영된다. 그러므로 예측 결과물의 에러가 10이 나온 것이 5로 나온 것보다, 정확히 2^2(4)배가 나쁜 도메인에서 쓰기 적합한 산식이다.
- 에러에 따른 손실이 기하 급수적으로 올라가는 상황에서 쓰기 적합하다. 

4.MAPE(Mean Absolute Percentage Error) 
- MAE를 퍼센트로 변환
- MAE와 같은 단점
- 모델에 대한 편향이 존재

```python
import datetime 

def dt_twohours_dt(str):
    dt = str
    dt_result = datetime.datetime(dt.year, dt.month, dt.day, 2*(dt.hour // 2), 60*(dt.minute // 60))
    return dt_result

def str_twohours_dt(str):
    dt = datetime.datetime.strptime(str, '%Y-%m-%d %H:%M:%S')
    dt_result = datetime.datetime(dt.year, dt.month, dt.day, 2*(dt.hour // 2), 0)
    return dt_result

def str_twohours_str():
    dt = datetime.datetime.strftime('%Y-%m-%d %H:%M:%S')
    dt_result = datetime.datetime(dt.year, dt.month, dt.day, 2*(dt.hour // 2), 60*(dt.minute // 60))
    return datetime.datetime.strftime(dt_result, '%Y-%m-%d %H:%M:%S')

start_end['new'] = start_end['시작시간'].apply(lambda dt: str_twohours_dt(dt))
```
